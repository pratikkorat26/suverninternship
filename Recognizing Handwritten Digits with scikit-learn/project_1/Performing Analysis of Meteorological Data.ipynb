# Needed imports for analysis

```python
#Pandas for the reading csv files and numpy for numerical ops and matplotlib for visualization of our data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1) Reading our dataset into Jupyter Notebook

```python
data = pd.read_csv("weatherHistory.csv", parse_dates = ['Formatted Date'] , index_col = ['Formatted Date'])
```

### 2) Inspecting our imported data

```python
# Inspecting features of our data
data.info()
```

By the above record we can see that most of our features get the float value and 3 of them get string values

```python
# Describing the dataset
data.describe()
```

This gives us statistics of our data like mean , standard deviation , min value and max value

```python
# Now we have to check how many records are invalid in sense of whether any of them having inapropriate value
# and any of them empty or not

# If there is null record or not
print(data.isnull().sum())
print("\n")
# If there is na value or not
print(data.isna().sum())
```

Look likes we got some invalid data so we have to remove unneccesary data

```python
# New data
# Removing na data
data = data.dropna()
```

```python
data.isnull().sum()

# Looks like we get rid of our na values data and be having clean data for further validation
```

```python
data.index = pd.to_datetime(data.index , utc =True)
```

```python
data.head()
```

# Using resampling function from pandas library

```python
data = data.resample('M').mean() # resample accroading to Month end ('M')
# This function will resample our dataframe by taking mean of every month in our dataset
```

```python
data.head()
```

```python
data.tail()
```

```python
#Here we have got the data from month to month starting from december of 2005 to end of 2016
```

```python
# Adding addtional data for visualization
data['month'] = data.index.month
data['year'] = data.index.year
data.index = data.index.date
```

```python
data.head()
```

```python
# YEar wise sampling
data_yearwise = data.resample("Y").mean()
```

```python
data_yearwise
```

# Visualization 

```python
# 1) line plot to get the hint how this data varying  from year to year
```

```python
def line_plotting_features():
    colors = ["blue", "green","red","cyan", "magenta","yellow","black" , "blue"]
    for i,column in enumerate(data_yearwise.columns[:-2]):
        plt.plot(data_yearwise[column] , color = colors[i])
        plt.legend(column)
        plt.xlabel(f"Year wise {column} data")
        plt.show()
```

```python
# Time series visualization , i mean how they changes over the times
line_plotting_features()
```

```python
# Bar plot for the same reason how they showing up in the given data
```

```python
def bar_plot_visualization():
    years = data_yearwise["year"].values
    for column in data_yearwise.columns[:-2]:
        sns.barplot(years , data_yearwise[column])
        plt.title(f"year wise {column}")
        plt.show()
```

```python
bar_plot_visualization()
```

```python
# We'll have look at correlation for the all the features
```

```python
# Compute the correlation matrix
corr = data_yearwise.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#Distribution plot 

def dist_plot_visualization():
    years = data_yearwise["year"].values
    for column in data_yearwise.columns[:-2]:
        sns.distplot(data_yearwise[column] , kde = True)
        plt.title(f"year wise {column}")
        plt.show()
```

```python
dist_plot_visualization()
```
